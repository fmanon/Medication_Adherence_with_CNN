```python
# Import libraries and modules
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.python.keras import utils
 
%matplotlib inline
```

```python
X = np.load('../data/X_train.npy', allow_pickle = True)
y = np.load('../data/y_train.npy', allow_pickle = True)
#validation_set = np.load('../data/z_test.npy', allow_pickle = True)
```

```python
X = X/255.0
y = np.array(y)
y = utils.to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
# Grid of Hyperparams
np.random.seed(42)
dense_layers = [1, 2, 3]
dense_neurons = [64, 128, 256]
conv_layers = [1, 2, 3]

for dense_layer in dense_layers:
    for layer_size in dense_neurons:
        for conv_layer in conv_layers:
            name = str(conv_layer) + '-conv, ' + str(layer_size) + ' nodes, ' + str(dense_layer) + ' dense'
            print(name)

            # Convolution and Pooling Layers
            model = Sequential()

            for _ in range(conv_layer):
                model.add(Conv2D(6, kernel_size = 3, activation = 'relu', input_shape=(X_train.shape[1:])))
                model.add(MaxPooling2D(pool_size=(2, 2)))

            model.add(Flatten())

            # Densely Connected Layers
            for _ in range(dense_layer):
                model.add(Dense(layer_size, activation = 'softmax'))

            # Output Layer
            model.add(Dense(y.shape[1], activation = 'softmax'))
           
            model.compile(loss='categorical_crossentropy',
                          optimizer='adam',
                          metrics=['accuracy'])

            history = model.fit(X_train, y_train,
                                batch_size= 128,
                                epochs=5,
                                validation_data = (X_test, y_test))
            
            print('***********************************************************************************************************************************************')
```

```python
X_train.shape[1:]
```

```python
# Best CNN: 2-conv, 64 nodes, 3 dense
# Grid of Hyperparams
np.random.seed(42)

# Convolution and Pooling Layers
model = Sequential()

# Layer 1
model.add(Conv2D(6, 
                 kernel_size = 3, 
                 activation = 'relu', 
                 input_shape=X_train.shape[1:]))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Layer 2
model.add(Conv2D(6, 
                 kernel_size = 3, 
                 activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
          
model.add(Flatten())

# Densely Connected Layers
model.add(Dense(64, activation = 'relu'))

# Output Layer
model.add(Dense(y.shape[1], activation = 'softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(X_train, y_train,
                    batch_size= 128,
                    epochs=5,
                    validation_data = (X_test, y_test))
```

```python
# Check out the model summary.
model.summary()
```

```python
# Check out our train loss and test loss over epochs.
train_loss = history.history['loss']
test_loss = history.history['val_loss']

# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_loss, label='Training Loss', color='#185fad')
plt.plot(test_loss, label='Testing Loss', color='orange')

# Set title
plt.title('Training  and Testing by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Categorical Crossentropy', fontsize = 18)
plt.xticks([0, .01, .02, .03, .04], [.01, .02, .03, .04, .05])

plt.legend(fontsize = 18);
```
